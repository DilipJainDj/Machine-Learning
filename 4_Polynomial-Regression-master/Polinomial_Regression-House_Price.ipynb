```python
import numpy as np
import matplotlib.pyplot as myPlot
import pandas as pd
```

```python
# Importing the dataset
myData = pd.read_csv('Housing_Price.csv')
myData
```

```python
X = myData.iloc[:, 0:1].values
X
```

```python
y = myData.iloc[:, 1].values
y
```

```python
myPlot.scatter(X,y, color='b')
myPlot.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
myModel_linear = LinearRegression().fit(X,y)
```

```python
myPlot.scatter(X,y, color='r')
myPlot.plot(X,myModel_linear.predict(X), color='b')
myPlot.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
X_poly = PolynomialFeatures(degree=3).fit_transform(X)
X_poly
```

```python
myModel_poly = LinearRegression().fit(X_poly,y)
```

```python
myPlot.scatter(X,y, color='r')
myPlot.scatter(X,myModel_poly.predict(X_poly), color='b')
myPlot.show()
```
